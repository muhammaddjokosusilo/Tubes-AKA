```python
import time
import matplotlib.pyplot as plt  # Untuk membuat grafik

class TravelBuddy:
    def __init__(self, hotels):
        self.hotels = sorted(hotels)  # Pastikan data terurut untuk binary search

    def binary_search_iterative(self, target):
        low, high = 0, len(self.hotels) - 1
        while low <= high:
            mid = low + (high - low) // 2
            if self.hotels[mid] == target:
                return mid
            elif self.hotels[mid] < target:
                low = mid + 1
            else:
                high = mid - 1
        return -1

    def binary_search_recursive(self, target, low, high):
        if low > high:
            return -1
        mid = low + (high - low) // 2
        if self.hotels[mid] == target:
            return mid
        elif self.hotels[mid] < target:
            return self.binary_search_recursive(target, mid + 1, high)
        else:
            return self.binary_search_recursive(target, low, mid - 1)

def measure_runtime(tb, target, search_method):
    start_time = time.time()  # Mulai pengukuran waktu
    if search_method == "iterative":
        tb.binary_search_iterative(target)
    else:
        tb.binary_search_recursive(target, 0, len(tb.hotels) - 1)
    end_time = time.time()  # Akhiri pengukuran waktu
    return end_time - start_time

def main():
    print("Ketua Kelompok : Trimarselinus Sitanggang(2311102209)")
    print("Anggota Kelompok : Muhammad Djoko Susilo(2311102212)")
    print("Studi Kasus:")
    print("TravelBuddy adalah aplikasi perjalanan yang bikin rencana liburan jadi gampang dan seru. Kamu bisa dapat panduan perjalanan yang disesuaikan dengan minatmu,")
    print("rekomendasi hotel dan restoran yang kece, plus fitur untuk bikin itinerary yang pas banget buat kamu. TravelBuddy bakal bikin setiap perjalananmu jadi pengalaman yang nggak terlupakan.")
    print("Kamu juga bisa gabung ke komunitas traveler buat berbagi tips dan cerita seru perjalanan.")
    print("Diaplikasi ini diperlukan database untuk menyimpan data dan dalam aplikasi ini memiliki fitur untuk mencari nilai dari database tersebut.")
    print("Dalam mencari data didalam aplikasi ini menggunakan metode binary search yang akan dihitung running time dari binary search yang menggunakan algoritma iteratif dan rekursif")
    print("--------------------------------------------------")
    print("Selamat Datang di Travel Buddy!")
    print("Fitur: Perhitungan running time dari banyaknya hotel yang diinputkan dan mencari nilai target untuk menjadi batas berhentinya perulangan")
    print("--------------------------------------------------")

    # Data untuk grafik
    data_sizes = []
    iterative_times = []
    recursive_times = []

    while True:
        # Input ukuran dataset dari pengguna
        try:
            data_size = int(input("Masukkan ukuran dataset (contoh: 10, 100, 1000): "))
            if data_size <= 0:
                raise ValueError("Ukuran dataset harus lebih besar dari 0.")
        except ValueError as e:
            print(f"Input tidak valid: {e}")
            continue

        # Tambahkan ukuran dataset ke daftar untuk grafik
        data_sizes.append(data_size)

        # Membuat dataset hotel
        dataset = [f"hotel-{i}" for i in range(data_size)]
        tb = TravelBuddy(dataset)

        # Target hotel yang tidak ada untuk pengukuran kasus terburuk
        target = f"hotel-{data_size + 100}"

        # Mengukur waktu untuk metode iteratif
        iterative_runtime = measure_runtime(tb, target, "iterative")
        iterative_times.append(iterative_runtime)

        # Mengukur waktu untuk metode rekursif
        recursive_runtime = measure_runtime(tb, target, "recursive")
        recursive_times.append(recursive_runtime)

        # Menampilkan hasil
        print("--------------------------------------------------")
        print("Hasil Pencarian:")
        print(f"Binary Search Iteratif: {iterative_runtime:.8f} detik")
        print(f"Binary Search Rekursif: {recursive_runtime:.8f} detik")

        # Membuat grafik runtime
        plt.figure(figsize=(10, 6))
        plt.plot(data_sizes, iterative_times, label="Iterative", marker='o')
        plt.plot(data_sizes, recursive_times, label="Recursive", marker='s')
        plt.title("Runtime Binary Search (Iterative vs Recursive)", fontsize=14)
        plt.xlabel("Ukuran Dataset", fontsize=12)
        plt.ylabel("Runtime (detik)", fontsize=12)
        plt.legend()
        plt.grid(True)
        plt.show()

        # Menanyakan apakah pengguna ingin melanjutkan
        print("--------------------------------------------------")
        choice = input("Apakah Anda ingin menambahkan ukuran dataset lain? (ya/tidak): ").strip().lower()
        if choice != 'ya':
            print("Menampilkan grafik runtime pencarian...")
            break



if __name__ == "__main__":
    main()
```
